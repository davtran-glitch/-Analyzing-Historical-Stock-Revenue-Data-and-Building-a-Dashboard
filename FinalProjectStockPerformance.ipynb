QUESTION 2 -- EXTRACTING TESLA REVENUE USING WEBSCRAPING

```python
# STEP 1 -- SEND AN HTTP REQUEST TO THE WEB PAGE
url = 'https://companiesmarketcap.com/aud/tesla/revenue/'
```

```python
data = requests.get(url).text
data
```

```python
# STEP 2 -- PARSE THE HTML CONTENT
```

```python
# Create a new BeautifulSoup object which pass 2 agruments (the HTML or XML content that you want to parse as a string, The name of the parser that you want to use to parse the HTML or XML content )
soup = BeautifulSoup(data, 'html5lib')
```

```python
tsla_revenue = pd.DataFrame(columns = ['Year', 'Annual Revenue'])
```

```python
for row in soup.find('tbody').find_all('tr'):
    col = row.find_all('td')
    year = col[0].text
    annual_revenue = col[1].text
    tsla_revenue = pd.concat([tsla_revenue, pd.DataFrame({'Year': [year], 'Annual Revenue': [annual_revenue]})], ignore_index = True)
```

```python
tsla_revenue['Annual Revenue'] = tsla_revenue['Annual Revenue'].str.extract(r'(\d+\.\d+|\d+)').astype(float)
tsla_revenue['Year'] = tsla_revenue['Year'].str.extract(r'(\d+)').astype(int)
```

```python
tsla_revenue.tail()
```

```python
#Question 4 -- EXTRACT GME REVENUE
gmeurl = 'https://companiesmarketcap.com/aud/gamestop/revenue/'
```

```python
gmereq = requests.get(gmeurl).text
gmereq
```

```python
gmesoup = BeautifulSoup(gmereq, 'html5lib')
```

```python
gme_revenue = pd.DataFrame(columns=['Year', 'Annual Revenue'])
```

```python
for row in gmesoup.find('tbody').find_all('tr'):
    col = row.find_all('td')
    year = col[0].text
    annual_revenue = col[1].text
    gme_revenue = pd.concat([gme_revenue, pd.DataFrame({'Year': [year], 'Annual Revenue': [annual_revenue]})], ignore_index = True)
```

```python
gme_revenue['Annual Revenue'] = gme_revenue['Annual Revenue'].str.extract(r'(\d+\.\d+|\d+)').astype(float)
```

```python
gme_revenue['Year'] = gme_revenue['Year'].str.extract(r'(\d+)').astype(int)
```

```python
gme_revenue.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def stock_dashboard(stock_data, stock_revenue, stock_name):
  fig,ax = plt.subplots(1,2, figsize = (22, 8))
  #Create line chart for stock price
  sns.lineplot(x=stock_data.Date, y=stock_data.Close, data=stock_data, ax=ax[0],  linewidth=2.5)
  ax[0].set_title(stock_name + ' Stock Price', fontsize=16, fontweight='bold')
  ax[0].set_xlabel('Date')
  ax[0].set_ylabel('Close Price (USD)')
  ax[0].grid(True, linestyle='--', alpha=0.7)
  ax[0].tick_params(axis='both', labelsize=10)
  ax[0].spines['top'].set_visible(False)
  ax[0].spines['right'].set_visible(False)
  #Create line chart for stock revenue
  sns.lineplot(x=stock_revenue.Year, y=stock_revenue['Annual Revenue'], data=stock_revenue, ax=ax[1],linewidth=2.5)
  ax[1].set_title(stock_name + ' Annual Revenue', fontsize=16, fontweight='bold')
  ax[1].set_xlabel('Year', fontsize=12)
  ax[1].set_ylabel('Annual Revenue (in Billion USD)', fontsize=12)
  ax[1].grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
  ax[1].tick_params(axis='both', which='major', labelsize=10)
  ax[1].spines['top'].set_visible(False)
  ax[1].spines['right'].set_visible(False)
```

```python
# QUESTION 5 -- TESLA STOCK AND REVENUE DASHBOARD
stock_dashboard(tesla_data, tsla_revenue,'TSLA')
```

```python
# QUESTION 6 -- GME STOCK AND REVENUE DASHBOARD
stock_dashboard(gme_data, gme_revenue,'GME')
```
